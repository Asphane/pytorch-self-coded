```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
```

```python
class CustomDataset(Dataset):
    def __init__(self, txt, labels):
        self.txt = txt
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        label = self.labels[idx]
        txt = self.txt[idx]
        sample = {'txt': txt, 'labels': label}
        return sample
```

```python
text = ['Happy', 'Amazing', 'Sad', 'Unhapy', 'Glum']
labels = ['Positive', 'Positive', 'Negative', 'Negative', 'Negative']

txt_labels_df = pd.DataFrame({'txt': text, 'labels': labels})
txt_labels_df.head()
```

```python
td = CustomDataset(txt_labels_df['txt'], txt_labels_df['labels'])
```

```python
dataloader = DataLoader(td, batch_size=2, shuffle=True)
```

```python
# Display text and label.
print('\nFirst iteration of data set: ', next(iter(td)), '\n')
# Print how many items are in the data set
print('Length of data set: ', len(td), '\n')
# Print entire data set
print('Entire data set: ', list(dataloader), '\n')
```
